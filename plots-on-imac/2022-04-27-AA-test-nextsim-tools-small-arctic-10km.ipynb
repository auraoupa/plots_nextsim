```python
import matplotlib.pyplot as plt

from pynextsim import NextsimBin

%matplotlib notebook
```

```python
n0 = NextsimBin('/Users/alberta/Data/nextsim/small_arctic_10km/experiment_062021/field_final.bin')
n0.variables
```

```python
n0.plot_var('Concentration')
```

```python
n0.plot_var('Thickness', cmap='ice', clim=[0, 0.5])
```

```python
# get coordinates of nodes
x, y = n0.mesh_info.get_nodes_xy()
# get indeces of coordinates for each element
indices = n0.mesh_info.get_indices() - 1
# get concentration for each element
c1 = n0.get_var('Concentration')
c2 = n0.get_var('Thickness')
```

```python
plt.figure()
plt.plot(x,y,'.')
plt.show()
```

```python
# shape of elements, nodes and indices
print(c1.shape, x.shape, indices.shape)
# x for 3 nodes of each elements
print(x[indices].shape)
```

```python
plt.figure()
plt.tripcolor(x, y, c1, triangles=indices, cmap='ice', clim=[0, 1])
plt.show()
```

```python
plt.figure()
plt.tripcolor(x, y, c1+c2, triangles=indices, cmap='ice', clim=[0, 1])
plt.colorbar()
plt.show()
```
